```python
#Pull Wine Data from local postgres DB
reds_df = pd.read_sql_table("red_wines", f'postgresql://{username}:{password}@localhost/{db_name}')
whites_df = pd.read_sql_table("white_wines", f'postgresql://{username}:{password}@localhost/{db_name}')
wine_df = pd.read_sql_table("wines", f'postgresql://{username}:{password}@localhost/{db_name}')
```

```python
wine_df.head()
```

```python
#This would be classifying them into written category names
#reds_df["Quality Category"]=np.where(reds_df['quality']>7,'Great',(np.where(reds_df['quality']<5,'Bad','Good')))

#Adding them to boolean category 1 == Good, 0 == Bad
reds_df['Quality Category']=np.where(reds_df['quality']>4,1,0)
```

```python
reds_df.head()
```

```python
target = reds_df['Quality Category']
target_names = ["Bad", "Good"]

red_data = reds_df.drop(['Quality Category','wine_type','quality'], axis=1)

feature_names = red_data.columns
red_data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(red_data, target, random_state=42)
```

```python
#Decision Tree Classifier
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
#Random Forrest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
